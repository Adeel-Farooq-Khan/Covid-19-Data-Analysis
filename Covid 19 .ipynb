```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.subplots import make_subplots
import plotly.express as px
from datetime import datetime
```

```python
covid_df = pd.read_csv('covid_19_india.csv')
covid_df.head(10)
```

```python
vaccine_df = pd.read_csv('covid_vaccine_statewise.csv')
vaccine_df.head(10)
```

```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace = True, axis = 1)
```

```python
covid_df.head()
```

```python
covid_df["Date"] = pd.to_datetime(covid_df["Date"],format = "%Y-%m-%d")
```

```python
covid_df.head()
```

```python
#Active Cases

covid_df['Active_Cases'] = covid_df['Confirmed'] - covid_df['Cured'] + covid_df["Deaths"]
```

```python
covid_df.tail()
```

```python
covid_df.head()
```

```python
covid_df.tail(10)
```

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths","Cured"],index = "State/UnionTerritory",aggfunc = max)
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by = "Confirmed" , ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# Top 10 active cases states

# Group by state and select the maximum active cases and date for each state
grouped_data = covid_df.groupby("State/UnionTerritory").agg({'Active_Cases': 'max', 'Date': 'max'}).reset_index()

# Sort the DataFrame by Active_Cases in descending order
sorted_data = grouped_data.sort_values(by="Active_Cases", ascending=False)

# Get the top 10 states with the highest active cases
top_10_active_cases = sorted_data.head(10)
```

```python
fig = plt.figure(figsize= (16,9))
```

```python
plt.figure(figsize=(16, 9))
plt.title("Top 10 States With Most Active Cases", size=25)

# Use seaborn to create the bar plot
ax = sns.barplot(data=top_10_active_cases, y="Active_Cases", x="State/UnionTerritory", linewidth=2, edgecolor="red")

# Set labels for x and y axes
plt.xlabel("States")
plt.ylabel("Active Cases")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

```python
#top states with highest deaths

top_10_deaths = covid_df.groupby("State/UnionTerritory").agg({'Deaths': 'max', 'Date': 'max'}).reset_index()

# Sort the DataFrame by Death in descending order
top_10_deaths = top_10_deaths.sort_values(by="Deaths", ascending=False)

plt.figure(figsize=(18, 5))
plt.title("Top 10 States with the Most Deaths", size=25)

# Use seaborn to create the bar plot
ax = sns.barplot(data=top_10_deaths.head(10), y="Deaths", x="State/UnionTerritory", linewidth=2, edgecolor="black")

# Set labels for x and y axes
plt.xlabel("States")
plt.ylabel("Total Death Cases")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

```python
# Filter the DataFrame to include data only for the selected states
selected_states = ['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh']
filtered_df = covid_df[covid_df['State/UnionTerritory'].isin(selected_states)]

plt.figure(figsize=(18, 8))
ax = sns.lineplot(data=filtered_df, x='Date', y='Active_Cases', hue='State/UnionTerritory')

plt.title("Top 5 Affected States", size=16)
plt.xlabel("Date")
plt.ylabel("Active Cases")



plt.show()
```

```python

vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine = vaccine.copy()
vaccine.columns = vaccine.columns.str.strip()

# Rename the column
vaccine.rename(columns={"Total Individuals Vaccinated": "Total"}, inplace=True)

# Display the updated DataFrame
vaccine.head()
```

```python
# Most vaccinated State

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending=False)[:5]

print(max_vac)
```

```python
plt.figure(figsize=(10, 5))
plt.title("Top 5 Vaccinated States", size=20)

x = sns.barplot(data=max_vac, y=max_vac['Total'], x=max_vac.index, linewidth=2, edgecolor="black")


plt.xlabel("States")
plt.ylabel("Vaccination")


plt.show()
```

```python
min_vaccine = max_vac.loc[max_vac['Total'].idxmin()]

print("State with the Minimum Vaccination:")
print(min_vaccine)
```
